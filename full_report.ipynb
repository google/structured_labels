```python
# --- stuff to import and show figures
import matplotlib.pyplot as plt
import pickle 

RESULTS_DIR = '/data/ddmg/slabs/waterbirds/results'

def show_figure(figure_file_name):  
    fig_handle = pickle.load(open(figure_file_name,'rb'))
    dummy = plt.figure()
    new_manager = dummy.canvas.manager
    new_manager.canvas.figure = fig_handle
    fig_handle.set_canvas(new_manager.canvas)
    fig_handle.show()
```

## Legend explanation

1. DNN --> simple "vanilla" baseline 
2. W-DNN --> baseline with weighting 
3. U-Ours --> DNN with MMD penalty, but no weighting 
4. W-Ours --> DNN with MMD penalty + weighting 
5. W-Ours (no bal) --> DNN with MMD penalty + weighting. Numerator of the weights is 1 not p(Y=y)


## Mean prediction broken up by group
- The off-diagonal plots here are the "important" ones
- All models give an incorrectly high (average) prediction for land birds on water backgrounds (top right).
- Similarly with water birds on land backgrounds: the average prediction of all the models is  low (bottom left) 
- The full weighted version of our model is suprisingly certain of its bad predictions. 


```python
show_figure(f'{RESULTS_DIR}/biased_balanced_class_mean_prediction_g.pkl')
```

## Aggregate accuracy

```python
show_figure(f'{RESULTS_DIR}/biased_balanced_class_accuracy.pkl')
```

## Accuracy broken down by group
- the two models that seem to be better in general are the unweighted model with the MMD penalty and the unbalanced weighting model with the MMD penalty
- predictions for water birds on land backgrounds are worse than land birds on water backgrounds, which might be beacause water birds are under-represented in the original data. 


```python
show_figure(f'{RESULTS_DIR}/biased_balanced_class_accuracy_g.pkl')
```
